```r
cities=names(data_json)
```

```r
# Initialize an empty list to store the records
lst <- list()
curr_itr=0
print(cities)
# Iterate through the keys (cities) in the data_json list
for (city in cities) {
    city_data <- data_json[[city]]
    curr_itr=curr_itr+1
    print(curr_itr)
    # Check if there is a link element and if there are restaurants in the city
    if (!is.null(city_data$link) && (!is.null(city_data$restaurants) || length(city_data$restaurants) == 0)) {
        # Initialize sub-city and sub-city link variables
        sub_city <- ""
        sub_city_link <- ""

        # Iterate through the restaurant codes in the city
        for (restau_code in names(city_data$restaurants)) {
            restaurant_data <- city_data$restaurants[[restau_code]]

            # Check the number of keys in the restaurant data
            if (length(names(restaurant_data)) != 5) {
                # Extract restaurant information
                restaurant <- paste0('"', restaurant_data$name, '"')
                rating <- restaurant_data$rating
                rating_count <- restaurant_data$rating_count
                cost <- restaurant_data$cost
                address <- paste0('"', gsub('"', "'", restaurant_data$address), '"')
                cuisine <- paste0('"', restaurant_data$cuisine, '"')
                licension_no <- restaurant_data$lic_no
                link <- restaurant_data$link

                # Iterate through the menus in the restaurant
                for (menu in names(restaurant_data$menu)) {
                    menu_ <- paste0('"', gsub('"', "'", menu), '"')

                    # Iterate through the items in the menu
                    for (item in names(restaurant_data$menu[[menu]])) {
                        item_ <- paste0('"', gsub('"', "'", item), '"')
                        price <- restaurant_data$menu[[menu]][[item]]$price
                        veg_or_non_veg <- restaurant_data$menu[[menu]][[item]]$veg_or_non_veg

                        # Create a record with all the extracted information
                        record <- c(city, city_data$link, sub_city, sub_city_link, restau_code, restaurant, rating)
                        record <- c(record, rating_count, cost, address, cuisine, licension_no, link, menu_, item_, price, veg_or_non_veg)

                        # Append the record to the list
                        lst <- append(lst, list(record))
                    }
                }
            } else {
                # If the keys length is 5, extract the appropriate data
                restaurant <- paste0('"', restaurant_data$name, '"')
                rating <- restaurant_data$rating
                rating_count <- ""
                cost <- ""
                address <- ""
                cuisine <- paste0('"', restaurant_data$type, '"')
                licension_no <- ""
                link <- restaurant_data$link
                menu <- ""
                item <- ""
                price <- restaurant_data$price
                veg_or_non_veg <- ""

                # Create a record with the extracted information
                record <- c(city, city_data$link, sub_city, sub_city_link, restau_code, restaurant, rating)
                record <- c(record, rating_count, cost, address, cuisine, licension_no, link, menu, item, price, veg_or_non_veg)

                # Append the record to the list
                lst <- append(lst, list(record))
            }
        }
    } else {
        # If there is no link or restaurants in the city, process sub-cities
        city_link <- city_data$link$link

        # Iterate through the sub-cities in the city
        for (sub_city in names(city_data)) {
            if (sub_city != "link") {
                sub_city_ <- paste0('"', gsub('"', "'", sub_city), '"')
                sub_city_link <- city_data[[sub_city]]$link

                # Iterate through the restaurant codes in the sub-city
                for (restau_code in names(city_data[[sub_city]]$restaurants)) {
                    restaurant_data <- city_data[[sub_city]]$restaurants[[restau_code]]

                    # Check the number of keys in the restaurant data
                    if (length(names(restaurant_data)) != 5) {
                        # Extract restaurant information
                        restaurant <- paste0('"', restaurant_data$name, '"')
                        rating <- restaurant_data$rating
                        rating_count <- restaurant_data$rating_count
                        cost <- restaurant_data$cost
                        address <- paste0('"', gsub('"', "'", restaurant_data$address), '"')
                        cuisine <- paste0('"', restaurant_data$cuisine, '"')
                        licension_no <- restaurant_data$lic_no
                        link <- restaurant_data$link

                        # Iterate through the menus in the restaurant
                        for (menu in names(restaurant_data$menu)) {
                            menu_ <- paste0('"', gsub('"', "'", menu), '"')

                            # Iterate through the items in the menu
                            for (item in names(restaurant_data$menu[[menu]])) {
                                item_ <- paste0('"', gsub('"', "'", item), '"')
                                price <- restaurant_data$menu[[menu]][[item]]$price
                                veg_or_non_veg <- restaurant_data$menu[[menu]][[item]]$veg_or_non_veg

                                # Create a record with all the extracted information
                                record <- c(city, city_link, sub_city_, sub_city_link, restau_code, restaurant, rating)
                                record <- c(record, rating_count, cost, address, cuisine, licension_no, link, menu_, item_, price, veg_or_non_veg)

                                # Append the record to the list
                                lst <- append(lst, list(record))
                            }
                        }
                    } else {
                        # If the keys length is 5, extract the appropriate data
                        restaurant <- paste0('"', restaurant_data$name, '"')
                        rating <- restaurant_data$rating
                        rating_count <- ""
                        cost <- ""
                        address <- ""
                        cuisine <- paste0('"', restaurant_data$type, '"')
                        licension_no <- ""
                        link <- restaurant_data$link
                        menu <- ""
                        item <- ""
                        price <- restaurant_data$price
                        veg_or_non_veg <- ""

                        # Create a record with the extracted information
                        record <- c(city, city_link, sub_city_, sub_city_link, restau_code, restaurant, rating)
                        record <- c(record, rating_count, cost, address, cuisine, licension_no, link, menu, item, price, veg_or_non_veg)

                        # Append the record to the list
                        lst <- append(lst, list(record))
                    }
                }
            }
        }
    }
}
print('Complete')
```

```r
# Convert the list of records to a data frame
df <- do.call(rbind.data.frame, lst)
colnames(df) <- c('city', 'city link', 'subcity', 'subcity link', 'restaurant code',
                  'restaurant', 'rating', 'rating count', 'cost', 'address',
                  'cuisine', 'licension no', 'restaurant link', 'menu', 'item',
                  'price', 'veg_or_non_veg')

# Write the data frame to a CSV file
write.csv(df, file = "Swiggy.csv", row.names = FALSE)
print('Complete')
```
