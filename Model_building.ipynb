```r
# Install and load necessary libraries
install.packages("FNN")
install.packages("ggplot2")
library(FNN)
library(ggplot2)

# Function to calculate sum of squared distances for each K
calculate_ssd <- function(train_data, max_k) {
  ssd <- numeric(max_k)
  for (k in 1:max_k) {
    knn_results <- try(get.knnx(train_data, train_data, k = k), silent = TRUE)
    if(inherits(knn_results, "try-error")) next  # Skip if error occurs
    ssd[k] <- sum(knn_results$nn.dist^2)
  }
  return(ssd)
}

# Increase the maximum value of K
max_k <- 30  # Maximum value of K to consider

# Calculate sum of squared distances for K values from 1 to max_k
ssd_values <- calculate_ssd(train_features, max_k)

# Plot the Elbow Method curve
elbow_df <- data.frame(K = 1:max_k, SSD = ssd_values)
ggplot(elbow_df, aes(x = K, y = SSD)) +
  geom_line(color = "blue") +
  geom_point(color = "red") +
  labs(title = "Elbow Method for KNN", x = "Number of Neighbors (K)", y = "Sum of Squared Distances") +
  theme_minimal()
```
